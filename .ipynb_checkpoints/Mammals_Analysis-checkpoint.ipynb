Investigating mammals sleep dataset from Open Intro after being inspired by listening to Matthew Walker's book Why We Sleep. Data download and more info can be found at - https://www.openintro.org/data/index.php?data=mammals

```python
# Data manipulation imports
import pandas as pd
import numpy as np

# Graphing imports
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Data/mammals.csv')
df.head()
```

```python
df.info()
```

# General Investigation

```python
sns.pairplot(df);
```

The pair plot above is tough to read with some outliers skewing a lot of the plots, nothing immediately insightful - main clear correlations are trivial like between total_sleep and dreaming and total_sleep with non_dreaming. Look to confirm with correlation heatmaps to non_dreaming, dreaming, and total_sleep:

```python
plt.figure(figsize = (3, 10))
ax = plt.axes()
ax.set_title('Feature Correlation to Total Sleep Time', {'fontsize' : 28})
corr = df.corr()[['total_sleep']]
sns.set(font_scale = 2)
sns.heatmap(corr.sort_values(by = 'total_sleep', ascending = False), 
            annot = True, cmap = 'mako', vmin = -1, vmax = 1, linecolor = 'black', 
            linewidth = 1, annot_kws = {'fontsize' : 16});
```

```python
plt.figure(figsize = (3, 10))
ax = plt.axes()
ax.set_title('Feature Correlation to Non Dreaming Time', {'fontsize' : 28})
corr = df.corr()[['non_dreaming']]
sns.set(font_scale = 2)
sns.heatmap(corr.sort_values(by = 'non_dreaming', ascending = False), 
            annot = True, cmap = 'mako', vmin = -1, vmax = 1, linecolor = 'black', 
            linewidth = 1, annot_kws = {'fontsize' : 16});
```

```python
plt.figure(figsize = (3, 10))
ax = plt.axes()
ax.set_title('Feature Correlation to Dreaming Time', {'fontsize' : 28})
corr = df.corr()[['dreaming']]
sns.set(font_scale = 2)
sns.heatmap(corr.sort_values(by = 'dreaming', ascending = False), 
            annot = True, cmap = 'mako', vmin = -1, vmax = 1, linecolor = 'black', 
            linewidth = 1, annot_kws = {'fontsize' : 16});
```

Interesting that total sleep as well as dreaming/non_dremaing are weakly negatively correlated with brain/body weight and life_span, try again when dropping outliers:

```python
df_do = df.loc[(df['brain_wt'] < 1000) & (df['body_wt'] < 1000) & (df['life_span'] < 60)].reset_index(drop = True)
df_do.head()
```

```python
sns.pairplot(df_do);
```

```python
plt.figure(figsize = (3, 10))
ax = plt.axes()
ax.set_title('Feature Correlation to Total Sleep Time (Dropped Outliers)', {'fontsize' : 28})
corr = df_do.corr()[['total_sleep']]
sns.set(font_scale = 2)
sns.heatmap(corr.sort_values(by = 'total_sleep', ascending = False), 
            annot = True, cmap = 'mako', vmin = -1, vmax = 1, linecolor = 'black', 
            linewidth = 1, annot_kws = {'fontsize' : 16});
```

```python
plt.figure(figsize = (3, 10))
ax = plt.axes()
ax.set_title('Feature Correlation to Non Dreaming Time (Dropped Outliers)', {'fontsize' : 28})
corr = df_do.corr()[['non_dreaming']]
sns.set(font_scale = 2)
sns.heatmap(corr.sort_values(by = 'non_dreaming', ascending = False), 
            annot = True, cmap = 'mako', vmin = -1, vmax = 1, linecolor = 'black', 
            linewidth = 1, annot_kws = {'fontsize' : 16});
```

```python
plt.figure(figsize = (3, 10))
ax = plt.axes()
ax.set_title('Feature Correlation to Dreaming Time (Dropped Outliers)', {'fontsize' : 28})
corr = df_do.corr()[['dreaming']]
sns.set(font_scale = 2)
sns.heatmap(corr.sort_values(by = 'dreaming', ascending = False), 
            annot = True, cmap = 'mako', vmin = -1, vmax = 1, linecolor = 'black', 
            linewidth = 1, annot_kws = {'fontsize' : 16});
```

Looks like total sleep and dream/non-dream time are actually negatively correlated with size and life span even when dropping outliers, confirm with heatmap:

```python
corr = df.corr()
plt.figure(figsize = (10, 10))
ax = plt.axes()
sns.set(font_scale = 1.2)
ax.set_facecolor('white') #set non-heatmap cell color to white, credit to stackover - https://stackoverflow.com/questions/52393084/how-to-assign-custom-color-to-masked-cells-in-seaborn-heatmap
sns.heatmap(corr, annot = True, cmap = 'Blues', ax = ax)
ax.set_title('Correlation to Total Sleep Time', {'fontsize' : 28})
plt.show()
```

Interesting that though body weight, brain weight, and life span are negatively correlated to sleep, non dreaming time, and dremaing time it is markedly less negative between brain weight and dreaming time

# Further Investigation
Dive into the correlations that aren't as intuitive - life span and total sleep are moderately negatively correlated, body weight and total sleep are moderately and negatively correlated, and that brain weight and dreaming time are moderately and negatively correlated. First investigate who has the highest life span at 100 and the highest body weights:

```python
df.sort_values(by = ['life_span'], ascending = False)
```

```python
df.sort_values(by = ['body_wt'], ascending = False).head(15)
```

```python
df.sort_values(by = ['brain_wt'], ascending = False).head(15)
```

Interesting that humans are listed as having a life span of 100, because that is way above all of the averages by country according to CIA data and many other reputable sources - https://www.cia.gov/the-world-factbook/field/life-expectancy-at-birth/country-comparison

Also there is no way the brain sizes are actually in kg as the data dictionary says, it is almost certainly meant to be in grams which is how I will reflect it below but is another consideration in the quality of this data:

```python
# Check food insecurity against density
plt.figure(figsize = (15, 8))
sns.regplot(data = df, x = 'total_sleep', y = 'life_span', scatter_kws = {'color': 'black'},
            line_kws = {'color': 'red'}, ci = None)

plt.title('Life Span Against Total Sleep For Mammals', size = 28)
plt.xlabel('Total Sleep (hours)', size = 22)
plt.ylabel('Life Span (years)', size = 22);
```

```python
# Check food insecurity against density
plt.figure(figsize = (15, 8))
g = sns.regplot(data = df, x = 'total_sleep', y = 'body_wt', scatter_kws = {'color': 'black'},
            line_kws = {'color': 'red'}, ci = None)

plt.title('Body Weight Against Total Sleep For Mammals', size = 28)
plt.xlabel('Total Sleep (hours)', size = 22)
plt.ylabel('Body Weight (kg)', size = 22);
```

```python
# Check food insecurity against density
plt.figure(figsize = (15, 8))
g = sns.regplot(data = df, x = 'dreaming', y = 'brain_wt', scatter_kws = {'color': 'black'},
            line_kws = {'color': 'red'}, ci = None)

plt.title('Brain Weight Against Dreaming Time For Mammals', size = 28)
plt.xlabel('Dreaming Time (hours)', size = 22)
plt.ylabel('Brain Weight (g)', size = 22);
```

There are no surprising correlations to dreaming, non-dreaming, or total sleep time, but it is interesting how life span and total sleep are moderately negatively correlated, body weight and total sleep are moderately and negatively correlated, and that brain weight and dreaming time are moderately and negatively correlated; all three of these correlations are counter-intuitive to me.
